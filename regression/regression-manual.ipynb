```python
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')
```

Data set for this exercise

```python
xs = np.array([1,2,3,4,5,6], dtype=np.float64)
ys = np.array([5,4,6,5,6,7], dtype=np.float64)

plt.scatter(xs,ys)
plt.show()
```

Manually calculating the slope by creating our own formula

```python
def best_fit_slope(xs,ys):
    m = ( (mean(xs) * mean(ys)) - mean(xs*ys) ) / ( (mean(xs)**2) - (mean(xs**2)) )
    return m
```

```python
best_fit_slope(xs,ys)
```

Creating function that will calculate slope and intercept

```python
def best_fit_slope_and_intercept(xs,ys):
    m = ( (mean(xs) * mean(ys)) - mean(xs*ys) ) / ( (mean(xs)**2) - (mean(xs**2)) )
    b = mean(ys) - m*mean(xs)
    return m, b
```

```python
m,b = best_fit_slope_and_intercept(xs,ys)
```

Creating regression line

```python
regression_line = [(m*x)+b for x in xs]
```

```python
plt.scatter(xs,ys)
plt.plot(xs, regression_line)
plt.show()
```

Adding a prediction

```python
predict_x = 8
predict_y = (m*predict_x)+b
```

```python
plt.scatter(xs,ys)
plt.scatter(predict_x,predict_y, color='green')
plt.plot(xs, regression_line)
plt.show()
```

Calculating R^2

```python
def squared_error(ys_orig, ys_line):
    return sum((ys_line-ys_orig)**2)

def coefficient_of_determination(ys_orig, ys_line):
    y_mean_line = [mean(ys) for y in ys_orig]
    squared_error_regr = squared_error(ys_orig, ys_line)
    squared_error_y_mean = squared_error(ys_orig, y_mean_line)
    return 1 - (squared_error_regr/squared_error_y_mean)
```

```python
coefficient_of_determination(ys, regression_line)
```
